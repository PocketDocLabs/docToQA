```python
from vllm_gen import get_completion_text
import re


context_split_prompt_file = "./prep_prompts/context-split-llama3.md"

# Test file
test_file = "./q-gen-test-doc.txt"

# Read in the test file
with open(test_file, "r") as f:
    test_text = f.read()

# Strip the test text
test_text = test_text.strip()

# Read in the context split prompt file
with open(context_split_prompt_file, "r") as f:
    context_split_prompt = f.read()

# New hotness
# Regex to match paragraphs between quadruple newlines
e_a_regex = test_text

# Replace groups of 2 or more newlines with a double newline
e_a_regex = re.sub(r"(\n){2,}", r"\n\n", e_a_regex)

# Replace ". " with "<DOT_SPACE>"
e_a_regex = re.sub("\. ", "<DOT_SPACE>", e_a_regex)

# Replace "? " with "<QUESTION_SPACE>"
e_a_regex = re.sub("\? ", "<QUESTION_SPACE>", e_a_regex)

# Replace "! " with "<EXCLAMATION_SPACE>"
e_a_regex = re.sub("\! ", "<EXCLAMATION_SPACE>", e_a_regex)

# Replace "\n\n" with "<DOUBLE_NEWLINE>"
e_a_regex = re.sub(r"\n\n", "<DOUBLE_NEWLINE>", e_a_regex)

# Replace "\n" with "<NEWLINE>"
e_a_regex = re.sub(r"\n", "<NEWLINE>", e_a_regex)

# Escape special characters
e_a_regex = re.escape(e_a_regex)

# Replace "<DOT_SPACE>" with ".(\n|\n\n| |\n</split>\n\n<split>\n)"
e_a_regex = re.sub(r"<DOT_SPACE>", r"\.(\\n|\\n\\n| |\\n</split>\\n\\n<split>\\n)", e_a_regex)

# Replace "<QUESTION_SPACE>" with "?(\n|\n\n| |\n</split>\n\n<split>\n)"
e_a_regex = re.sub(r"<QUESTION_SPACE>", r"\?(\\n|\\n\\n| |\\n</split>\\n\\n<split>\\n)", e_a_regex)

# Replace "<EXCLAMATION_SPACE>" with "!(\n|\n\n| |\n</split>\n\n<split>\n)"
e_a_regex = re.sub(r"<EXCLAMATION_SPACE>", r"\!(\\n|\\n\\n| |\\n</split>\\n\\n<split>\\n)", e_a_regex)

# Replace "<DOUBLE_NEWLINE>" with "(\n|\n\n| |\n</split>\n\n<split>\n)"
e_a_regex = re.sub(r"<DOUBLE_NEWLINE>", r"(\\n|\\n\\n| |\\n</split>\\n\\n<split>\\n)", e_a_regex)

# Replace "<NEWLINE>" with "(\n|\n\n| |\n</split>\n\n<split>\n)"
e_a_regex = re.sub(r"<NEWLINE>", r"(\\n|\\n\\n| |\\n</split>\\n\\n<split>\\n)", e_a_regex)

# Add the split tags to the beginning and end
e_a_regex = "<split>\\n" + e_a_regex + "\\n</split>"

# old and busted
# Regex to match paragraphs between quadruple newlines
# e_a_regex = """(<split>\n(.|\s)*?\n</split>\n\n)+"""

print(e_a_regex)

# Replace {{text}} with the test text
context_split_prompt = context_split_prompt.replace("{{text}}", test_text)

input = context_split_prompt

output = get_completion_text(input, regex=e_a_regex, max_tokens=5000, temperature=0.5, min_p=0.08, ignore_eos=True, skip_special_tokens=True)

# Write the output to a file
with open("output.txt", "w") as f:
    f.write(output)

```
