```python
from lcpp_gen import get_completion_text
import re


context_split_prompt_file = "./prep_prompts/context-split-llama3.md"

# Test file
test_file = "./MCTP 12-10C-ch1.md"

# Read in the test file
with open(test_file, "r") as f:
    test_text = f.read()

# Strip the test text
test_text = test_text.strip()

# Read in the context split prompt file
with open(context_split_prompt_file, "r") as f:
    context_split_prompt = f.read()


# old and busted
# Regex to match paragraphs between quadruple newlines
# e_a_regex = """(<split>\n(.|\s)*?\n</split>\n\n)+"""

# New hotness
# Regex to match paragraphs between quadruple newlines
grammar = test_text

# Escape any quotes
grammar = grammar.replace("\"", "\\\"")

# Remove any ebnf symbols (e.g. |, ""::=, +, *, ?)
# Split the regex into lines
grammar = grammar.split("\n")

# Replace whitespace with a space
grammar = [re.sub(r"\s+", " ", line) for line in grammar]

# Wrap each line in quotes
grammar = [f"\"{line}\"" for line in grammar]

# Join the lines with (\n|"</split>\n\n<split>")
grammar = "(\"\\n\"|\"\\n</split>\\n\\n<split>\\n\")".join(grammar)

# Add the start and end <split> tags
grammar = "root ::= \"<split>\\n\"" + grammar + "\"\\n</split>\""

print(grammar)

# Replace {{text}} with the test text
context_split_prompt = context_split_prompt.replace("{{text}}", test_text)

input = context_split_prompt

output = get_completion_text(input, grammar=grammar, max_tokens=15000, temperature=0.1, min_p=0.1, ignore_eos=True)

# Write the output to a file
with open("output.txt", "w") as f:
    f.write(output)

```
