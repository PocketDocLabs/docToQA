```python
import re
from backend.claude import get_completion_text



prompt_file = "/run/media/pocketdoc/Storage/datasets/tools/docToQA/generation_prompts/a-gen-mistral.txt"

with open(prompt_file, "r") as f:
    prompt = f.read()

notes_file = "/run/media/pocketdoc/Storage/datasets/tools/docToQA/notes.md"

with open(notes_file, "r") as f:
    notes = f.read()

question = "Let me know if you have any tips for identifying the subtle differences between similar opaque material like jasper versus petrified wood."

summary = "Washington's \"Rockhounding Washington\" guide, authored by Lars W. Johnson and published in 2018, lists 60 locations with maps and directions, updated in 2024 for collecting. Notable finds include blue beryl at Winchester Creek, red carnelian, jasper, agate, petrified wood, Eocene plant fossils, minerals, and gold. Accessible sites year round with seasonal restrictions. Top rock shops and museums include Ellensburg, Kittitas County Historical Museum, Burke Museum of Natural History, and Ginkgo Petrified Forest State Park. Author shares over a decade of experiences and tips. Clubs include NW Rockhounds, offering camps and BBQs, and Grants Harbor Geology, Mount Baker clubs nearby."


# Replace {{text}} in the prompt with the notes
prompt = re.sub(r"{{text}}", notes, prompt)

# Replace {{summary}} in the prompt with the summary
prompt = re.sub(r"{{summary}}", summary, prompt)

# Replace {{question}} in the prompt with the question
prompt = re.sub(r"{{question}}", question, prompt)

# Save the prompt to a file
with open("prompt.txt", "w") as f:
    f.write(prompt)

# print(prompt)

# Regex for a string of text ending at the first occurrence of "#### Warnings and disclaimers:"
regex = r"((\n|.)*?)####"

# Get the completion text
output = get_completion_text(prompt, max_tokens=1024, min_p=0.4, temperature=1.0, repetition_penalty=1.0, regex=regex, stop_sequence=["####"], model="claude-3-opus-20240229")

print(output)
```
